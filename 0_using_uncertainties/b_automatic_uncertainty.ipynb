## B: Automatic uncertainty monitoring in metatomic calculators

```python
import numpy as np
import ase.build
import ase.units
from ase.md import VelocityVerlet
from metatomic.torch.ase_calculator import MetatomicCalculator

np.random.seed(42)
```

When using `metatomic` calculators, users are displayed warnings of high-uncertainty inferences during a simulation when using uncertainty-enabled models. This feature is also available in the `metatomic` interfaces to other simulation engines. We demonstrate this as follows:

```python
# Build a fcc copper structure
copper = ase.build.bulk("Cu", "fcc", cubic=True)

# Initialize an ASE calculator with a pre-trained potential
calculator = MetatomicCalculator("pet-mad.pt")

# Set the calculator for the copper structure
copper.calc = calculator

# Initial small rattle to avoid zero forces
copper.rattle(0.1)

# Run some molecular dynamics
dyn = VelocityVerlet(copper, timestep=1.0 * ase.units.fs)
dyn.run(100)
print("First MD run completed!")
```

Let's now try and see what happens when we rattle the structure significantly:

```python
# Run again, this time with a crazy structure
copper.rattle(5.0)
dyn.run(100)
print("Second MD run completed (hopefully with a warning!)")
```

You can also see what happens when you change the `uncertainty_threshold` argument to the `metatomic` calculator (which defaults to 0.1 eV).
